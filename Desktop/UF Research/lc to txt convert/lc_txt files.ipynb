```python
import numpy as np 
from astropy.table import Table
from astropy.io import fits
from astropy.io import ascii
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('ls5039_lc1.txt', unpack=True).T

mjd = ((data[:,0]))
counts = ((data[:,1]))
error = (data[:,2])
ref_counts = (data[:,3])
ref_error = ((data[:,4]))

col1 = mjd
col2 = counts
col3 = error
col4 = ref_counts
col5 = ref_error

with open('test.txt', 'w') as f:
    f.write('{a:^8}{b:^8}{c:^8}{d:^8}{e:^8}'.format(a='\tMJD\t\t', b='COUNTS\t\t', c='ERROR\t    ',d='REF_COUNTS\t    ',e='REF_ERROR\n'))
    for a, b, c, d, e in zip(col1, col2, col3, col4, col5):
        f.write('{0:15}{1:15}{2:15}{3:15}{4:15}\n'.format(a, b, c, d, e))
```

```python
token = open('test.txt','r')
linestoken = token.readlines()
tokens_column_number = 0 # MJD 
mjd_values = []

for x in linestoken:
    mjd_values.append(x.split()[tokens_column_number])
    
tokens_column_number = 1 # Counts 
counts_values = []

for x in linestoken:
    counts_values.append(x.split()[tokens_column_number])
    
tokens_column_number = 2 # Errors 
error_values = []

for x in linestoken:
    error_values.append(x.split()[tokens_column_number])
    


tokens_column_number = 3 # Reference Counts  
refcount_values = []

for x in linestoken:
    refcount_values.append(x.split()[tokens_column_number])
    


tokens_column_number = 4 # Reference Errors   
referror_values  = []

for x in linestoken:
    referror_values.append(x.split()[tokens_column_number])
```

```python
# converting from string to float:

mjdfix = (mjd_values[1:]) 
countsfix =  (counts_values[1:]) 
errorsfix =  (error_values[1:]) 
refcountsfix =  (refcount_values[1:]) 
referrorsfix =  (referror_values[1:])

mjdnew = []
for i in mjdfix:
    item = float(i)
    mjdnew.append(item)

countsfixnew = [] 
for i in countsfix:
    item = float(i)
    countsfixnew.append(item)
    
errorsfixnew = []
for i in errorsfix:
    item = float(i)
    errorsfixnew.append(item)

refcountsfixnew = []
for i in refcountsfix:
    item = float(i)
    refcountsfixnew.append(item)
    
referrorsfixnew = []
for i in referrorsfix:
    item = float(i)
    referrorsfixnew.append(item)

mjd = mjdnew
counts = countsfixnew
errors = errorsfixnew
refcounts = refcountsfixnew
referrors = referrorsfixnew
```

```python
fig, ax = plt.subplots()
ax.plot(mjd, counts, '-o', label='Counts')
ax.plot(mjd, errors, '-b', label='Errors') 

legend = ax.legend(loc='upper right', shadow=False, fontsize='x-large')

fig = plt.gcf()
fig.set_size_inches(20,10)

plt.title("MJD vs Counts and Error for ls5039_lc1", fontsize = 25) # !!!
plt.xlabel("MJD", fontsize = 20)
plt.ylabel("Counts", fontsize = 20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylim(175000,220000)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(mjd, refcounts, '-o', label='Counts')
ax.plot(mjd, referrors, '-b', label='Errors') 

legend = ax.legend(loc='upper right', shadow=False, fontsize='x-large')

fig = plt.gcf()
fig.set_size_inches(20,10)

plt.title("MJD vs Counts and Error for ls5039_lc2", fontsize = 25) # !!!
plt.xlabel("MJD", fontsize = 20)
plt.ylabel("Counts", fontsize = 20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
# plt.ylim(170000,230000)
plt.show()
```
